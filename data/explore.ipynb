```python
import pandas as pd
from random import randint

df = pd.read_pickle('math_articles.pd')
articles = [df.iloc[[randint(0, len(df))]].article.values[0] for i in range(10)]
article = articles[0]
```

```python
import re

def get_colocations(regex, s):
    return [(m.start(), m.end()) for m in re.finditer(regex, s)]

# returns a list of start/end indices for math markup
def get_math_locales(txt):
    locales = []  
    bracket_math = get_colocations('{{math.+?}}', txt)
    tag_math = get_colocations('<math>.+?</math>', txt)
    quote_math = get_colocations("\\'\\'.+?\\'\\'", txt)
    return sorted(bracket_math + tag_math + quote_math)

l = get_math_locales(article)
l
```

```python
def get_context_windows(txt, colos, size=50):
    return [txt[(colo[0] - size): (colo[1] + size)] for colo in colos]   
    
get_context_windows(article, l)
```
